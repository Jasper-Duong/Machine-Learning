````python
import os
import tarfile
from six.moves import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing") #datasets/housing
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH): 
    #Create dir "datasets/housing/housing.tgz"
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    
    #Download dataset to tgz_path
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)

    #Open & Extract .tgz file
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
````

````python
import numpy as np

def split_train_test(data, test_ratio):
    
    np.random.seed(42)
    shuffled_indices = np.random.permutation(len(data)) 
    test_set_size = int(len(data) * test_ratio) 
    test_indices = shuffled_indices[:test_set_size] 
    train_indices = shuffled_indices[test_set_size:]
    
    return data.iloc[train_indices], data.iloc[test_indices]

train_data, test_data = split_train_test(df, 0.2)
train_data
````

Although setting random seed will guarantee the same split of the CURRENT Dataset but if the next time I split data on an updated set, the instances that used to be in test would not remained the same.

=> Need an identifier that's unique and immutable for each instance => Hash

````python
from zlib import crc32

def test_set_check(identifier, test_ratio): #identifier will be df's ID => Make sure new data get APPENDED , no rows DELETED
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32 

def split_train_test_by_id(data, test_ratio, column_name):
    ids = data[column_name]

    #check if id in test_set => True/False
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    
    return data.loc[~in_test_set], data.loc[in_test_set]
````

### Custom Transformer

````python
from sklearn.base import TransformerMixin, BaseEstimator
````

````python
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room 
    def fit(self, X, y=None):
        return self # nothing else to do 
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix] 
        population_per_household = X[:, population_ix] / X[:, households_ix] 
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
````

### Data Scaling & Transformation Pipelines

````python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('attr_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler())
])

housing_num_trans = num_pipeline.fit_transform(housing_num)
````

- Full PipeLine for bot Num/Cat attributes
    - ```ColumnTransformer```: apply appropriate transformations to corresponding columns.

````python
from sklearn.compose import ColumnTransformer

num_attr = list(housing_num)
cat_attr = ['ocean_proximity']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attr),
    ('cat', OneHotEncoder(), cat_attr)
])

housing_prepared = full_pipeline.fit_transform(housing)
````

### Decision Tree Regression

````python
from sklearn.tree import DecisionTreeRegressor

#Train
tree_reg = DecisionTreeRegressor()
tree_reg.fit(housing_prepared, df_label)

#Predictions => Evaluate
preds = tree_reg.predict(housing_prepared)
tree_mse = mean_squared_error(df_label,preds)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
````

**No error at all? Absolutely perfect?**
- Badly **OVERFITTING** data? 
> Split Training -> Train/Validation sets
>> **K-fold-CrossValidation**

````python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, housing_prepared, df_label,
                        scoring='neg_mean_squared_error', cv=10)

tree_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print(f"Scores: {scores}")
    print(f"Mean: {scores.mean()}")
    print(f"Standard Deviation: {scores.std()}") #Measure how precise the evaluation is
    
display_scores(tree_rmse_scores)
````

````python
lin_scores = cross_val_score(lin_reg, housing_prepared, df_label,
                            scoring='neg_mean_squared_error', cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
````

### Random Forest Regressor

- Train many DecisionTrees on **random subsets of features**.
- Average the predictions.
- ***Ensemble Learning***: Building a model on top of many other models.

````python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(housing_prepared, df_label)

forest_scores = cross_val_score(forest_reg, housing_prepared, df_label,
                               scoring='neg_mean_squared_error', cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
````

````python
forest_preds = forest_reg.predict(housing_prepared)
forest_rsme = np.sqrt(mean_squared_error(forest_preds, df_label))
print("RSME on Training set: ", forest_rsme)
````

- Score on Val (50K) vs Train (18K): Large margin

=> <strong> Overfitting on Train </strong>

- Solutions for Overfitting:
    - Simplify model.
    - Constrain (Regularize) model.
    - Get more training data.

## Evaluate on Test Set

````python
final_model = grid_search.best_estimator_

X_test = strat_test_set.drop("median_house_value", axis=1)
y_test = strat_test_set['median_house_value'].copy()

#Transform test data, NOT fit_transform() -> Don't want model to fit on test set
X_test_prepared = full_pipeline.transform(X_test)

final_predictions = final_model.predict(X_test_prepared)

final_rmse = np.sqrt(mean_squared_error(y_test, final_predictions))

print(f"RMSE on Test set: {final_rmse}")
````

- A performance score (generalization error) is not enough to express how precise the estimations are.

<strong> => 95% Confidence interval </strong>

````python
from scipy import stats

confidence = 0.95
squared_error = (final_predictions - y_test)**2
np.sqrt(stats.t.interval(confidence, len(squared_error) - 1,
                         loc=squared_error.mean(),
                         scale=stats.sem(squared_error)
))
````
